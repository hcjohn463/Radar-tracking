```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import hdbscan
from scipy.ndimage import label

# === 讀取 RDI 函數 ===
def read_h5_data(file_path):
    with h5py.File(file_path, 'r') as file:
        rdi_data = file['DS1'][0]
    return rdi_data

# CFAR 參數
def ca_cfar_2d(rdm, guard_rows, guard_cols, ref_rows, ref_cols, scale_factor):
    # ... (之前的 ca_cfar_2d_corrected 函數代碼) ...
    rows, cols = rdm.shape
    binary_map = np.zeros_like(rdm, dtype=np.uint8)
    threshold_map = np.full_like(rdm, np.inf, dtype=float) # 可選：用於除錯

    win_r_half = guard_rows + ref_rows
    win_c_half = guard_cols + ref_cols
    guard_r_half = guard_rows
    guard_c_half = guard_cols

    pad_width = ((win_r_half, win_r_half), (win_c_half, win_c_half))
    padded_rdm = np.pad(rdm, pad_width, mode='edge') # 使用 'edge' 填充通常比 0 好

    for r in range(rows):
        for c in range(cols):
            r_pad = r + win_r_half
            c_pad = c + win_c_half
            window = padded_rdm[r_pad - win_r_half : r_pad + win_r_half + 1,
                                c_pad - win_c_half : c_pad + win_c_half + 1]
            guard_mask = np.zeros_like(window, dtype=bool)
            guard_r_center = win_r_half
            guard_c_center = win_c_half
            guard_mask[guard_r_center - guard_r_half : guard_r_center + guard_r_half + 1,
                       guard_c_center - guard_c_half : guard_c_center + guard_c_half + 1] = True
            ref_values = window[~guard_mask]
            N = len(ref_values)
            if N == 0:
                threshold = np.inf
            else:
                background_estimate = np.mean(ref_values)
                threshold = background_estimate * scale_factor
            threshold_map[r, c] = threshold
            if rdm[r, c] > threshold:
                binary_map[r, c] = 1
    return binary_map
    
def filter_clusters(binary_map, min_size=4):
    labeled, num = label(binary_map)
    for i in range(1, num + 1):
        if np.sum(labeled == i) < min_size:
            binary_map[labeled == i] = 0
    return binary_map



# === 讀取資料與設定 ===
file_path = "bin1.h5"
rdi_data = read_h5_data(file_path)
start_value = 0
end_value = 5 # 注意：range() 的終止值不包含在內，所以要加 1
frame_ids = list(range(start_value, end_value))
range_bins = rdi_data.shape[0]
velocity_bins = rdi_data.shape[1]
x = np.linspace(-0.15, 0.15, velocity_bins)
y = np.linspace(0, 35, range_bins)

# === 圖片與 clustering 結果整合畫出來 ===
fig, axs = plt.subplots(len(frame_ids), 2, figsize=(10, 4 * len(frame_ids)))

for idx, frame_id in enumerate(frame_ids):
    rdm = rdi_data[:, :, frame_id]
    rdm_f32 = np.array(rdm, dtype=np.float32)

    binary_map = ca_cfar_2d(rdm_f32, 6, 6, 5, 5, 9)

    # Step 2: 濾掉雜訊 cluster
    cleaned = filter_clusters(binary_map, min_size= 6)
    coords = np.column_stack(np.where(cleaned == 1))

    # 圖 1：RDM
    axs[idx, 0].imshow(rdm_f32, extent=[x[0], x[-1], y[-1], y[0]], aspect='auto', cmap='jet')
    axs[idx, 0].set_title(f'RDM Frame {frame_id}')
    axs[idx, 0].set_xlabel('Velocity (cm/s)')
    axs[idx, 0].set_ylabel('Range (cm)')


    # 圖 3：DBSCAN
    axs[idx, 1].imshow(binary_map, extent=[x[0], x[-1], y[-1], y[0]], aspect='auto', cmap='hot')
    axs[idx, 1].set_title(f'DBSCAN Frame {frame_id}')
    axs[idx, 1].set_xlabel('Velocity (cm/s)')
    axs[idx, 1].set_ylabel('Range (cm)')
    if len(coords) > 0:
        clusterer = DBSCAN(eps=2, min_samples=3)
        labels = clusterer.fit_predict(coords)
        for cluster_id in set(labels):
            if cluster_id == -1:
                continue
            cluster_points = coords[labels == cluster_id]
            r_mean, d_mean = cluster_points.mean(axis=0)
            axs[idx, 1].plot(x[int(d_mean)], y[int(r_mean)], 'go')

plt.tight_layout()
plt.show()
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import hdbscan

# === 讀取 RDI 函數 ===
def read_h5_data(file_path):
    with h5py.File(file_path, 'r') as file:
        rdi_data = file['DS1'][0]
    return rdi_data

# === CA-CFAR 函數 ===
def ca_cfar_2d(rdm, guard_cells=(2, 2), reference_cells=(2, 2), scale_factor=4.0):
    rows, cols = rdm.shape
    g_row, g_col = guard_cells
    r_row, r_col = reference_cells

    binary_map = np.zeros_like(rdm, dtype=np.uint8)
    win_row = g_row + r_row
    win_col = g_col + r_col

    for i in range(win_row, rows - win_row):
        for j in range(win_col, cols - win_col):
            cut_value = rdm[i, j]
            r_start, r_end = i - win_row, i + win_row + 1
            c_start, c_end = j - win_col, j + win_col + 1

            ref_window = rdm[r_start:r_end, c_start:c_end].copy()
            ref_window[r_row:-r_row, r_col:-r_col] = 0

            ref_cells = ref_window[ref_window > 0]
            if ref_cells.size == 0:
                continue

            noise_avg = np.mean(ref_cells)
            threshold = noise_avg * scale_factor
            if cut_value > threshold:
                binary_map[i, j] = 1

    return binary_map

def os_cfar_2d(rdm, guard_cells=(2, 2), reference_cells=(5, 5), rank=10, scale_factor=1.5):
    """
    OS-CFAR：使用排序統計法設定門檻，選第 k 大的參考單元作為門檻。
    
    rdm: 2D Range-Doppler Map (float32)
    guard_cells: (g_row, g_col)
    reference_cells: (r_row, r_col)
    rank: 排序後取第幾大的作為門檻 (越小越容易偵測)
    scale_factor: 放大倍數（類似 CA-CFAR 的乘法門檻）
    """
    rows, cols = rdm.shape
    g_row, g_col = guard_cells
    r_row, r_col = reference_cells

    binary_map = np.zeros_like(rdm, dtype=np.uint8)

    win_row = g_row + r_row
    win_col = g_col + r_col

    for i in range(win_row, rows - win_row):
        for j in range(win_col, cols - win_col):
            # 擷取完整窗口區域
            r_start = i - win_row
            r_end   = i + win_row + 1
            c_start = j - win_col
            c_end   = j + win_col + 1

            window = rdm[r_start:r_end, c_start:c_end].copy()

            # 設定 guard cells + CUT 為 0
            gc_r_start = win_row - g_row
            gc_r_end   = win_row + g_row + 1
            gc_c_start = win_col - g_col
            gc_c_end   = win_col + g_col + 1
            window[gc_r_start:gc_r_end, gc_c_start:gc_c_end] = 0

            # 把有效參考值取出並排序
            ref_values = window[window > 0]
            if len(ref_values) < rank:
                continue  # 排名不夠就略過

            threshold = np.sort(ref_values)[-rank] * scale_factor

            if rdm[i, j] > threshold:
                binary_map[i, j] = 1

    return binary_map


# === 讀取資料與設定 ===
file_path = "hand.h5"
rdi_data = read_h5_data(file_path)
frame_ids = [0, 1, 2, 3, 4, 5]
range_bins = rdi_data.shape[0]
velocity_bins = rdi_data.shape[1]
x = np.linspace(-0.15, 0.15, velocity_bins)
y = np.linspace(0, 35, range_bins)

# === 圖片與 clustering 結果整合畫出來 ===
fig, axs = plt.subplots(len(frame_ids), 3, figsize=(15, 4 * len(frame_ids)))

for idx, frame_id in enumerate(frame_ids):
    rdm = rdi_data[:, :, frame_id]
    rdm_f32 = np.array(rdm, dtype=np.float32)
    binary_map = os_cfar_2d(
        rdm,
        guard_cells=(3, 3),
        reference_cells=(5, 5),
        rank=50,
        scale_factor=3
    )
    # binary_map = ca_cfar_2d(rdm_f32, guard_cells=(2,2), reference_cells=(7,7), scale_factor=3)
    coords = np.column_stack(np.where(binary_map == 1))

    # 圖 1：RDM
    axs[idx, 0].imshow(rdm_f32, extent=[x[0], x[-1], y[-1], y[0]], aspect='auto', cmap='jet')
    axs[idx, 0].set_title(f'RDM Frame {frame_id}')
    axs[idx, 0].set_xlabel('Velocity (cm/s)')
    axs[idx, 0].set_ylabel('Range (cm)')

    # 圖 2：模擬 HDBSCAN
    axs[idx, 1].imshow(binary_map, extent=[x[0], x[-1], y[-1], y[0]], aspect='auto', cmap='hot')
    axs[idx, 1].set_title(f'HDBSCAN Frame {frame_id}')
    axs[idx, 1].set_xlabel('Velocity (cm/s)')
    axs[idx, 1].set_ylabel('Range (cm)')
    if len(coords) > 0:
        clusterer = hdbscan.HDBSCAN(min_cluster_size=2, min_samples=1)
        labels = clusterer.fit_predict(coords)
        for cluster_id in set(labels):
            if cluster_id == -1:
                continue
            cluster_points = coords[labels == cluster_id]
            r_mean, d_mean = cluster_points.mean(axis=0)
            axs[idx, 1].plot(x[int(d_mean)], y[int(r_mean)], 'bo')

    # 圖 3：DBSCAN
    axs[idx, 2].imshow(binary_map, extent=[x[0], x[-1], y[-1], y[0]], aspect='auto', cmap='hot')
    axs[idx, 2].set_title(f'DBSCAN Frame {frame_id}')
    axs[idx, 2].set_xlabel('Velocity (cm/s)')
    axs[idx, 2].set_ylabel('Range (cm)')
    if len(coords) > 0:
        clusterer = DBSCAN(eps=1.5, min_samples=2)
        labels = clusterer.fit_predict(coords)
        for cluster_id in set(labels):
            if cluster_id == -1:
                continue
            cluster_points = coords[labels == cluster_id]
            r_mean, d_mean = cluster_points.mean(axis=0)
            axs[idx, 2].plot(x[int(d_mean)], y[int(r_mean)], 'go')

plt.tight_layout()
plt.show()
```

```python
#第 3 步：使用 clustering
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

# 找出所有 "亮點" 的座標 (row = range bin, col = doppler bin)
coords = np.column_stack(np.where(binary_map == 1))

# 建立 DBSCAN 分群模型
clusterer = DBSCAN(eps=1.5, min_samples=4)
labels = clusterer.fit_predict(coords)

# 找有幾群（不包含 label=-1 的雜訊）
num_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print(f"偵測到 {num_clusters} 個目標")

# 把每一群的座標算出中心點
for cluster_id in range(num_clusters):
    cluster_points = coords[labels == cluster_id]
    r_mean, d_mean = cluster_points.mean(axis=0)
    print(f"目標 {cluster_id+1}: range_bin={r_mean:.2f}, doppler_bin={d_mean:.2f}")

plt.imshow(binary_map, cmap='hot')
plt.title(f"DBSCAN clusters (Frame {frame_id})")
for cluster_id in range(num_clusters):
    cluster_points = coords[labels == cluster_id]
    r_mean, d_mean = cluster_points.mean(axis=0)
    plt.plot(d_mean, r_mean, 'bo')  # doppler 是橫軸，range 是縱軸
plt.colorbar()
plt.show()
```

```python
#第 4 步：從每個 cluster 提取 range 與 doppler 資訊

# 基本資訊
h, w = binary_map.shape
range_resolution = 35 / h             # cm per range bin
doppler_resolution = 0.3 / w          # cm/s per doppler bin
doppler_center = w // 2

for cluster_id in range(num_clusters):
    cluster_points = coords[labels == cluster_id]
    r_bin, d_bin = cluster_points.mean(axis=0)
    real_range = r_bin * range_resolution
    real_speed = (d_bin - doppler_center) * doppler_resolution
    print(f"目標 {cluster_id+1}: 距離 ≈ {real_range:.2f} cm, 速度 ≈ {real_speed:.4f} cm/s")
```
